```python
import pandas as pd
```

```python
# 1.	Load the retail_sales_dataset.csv file into a Pandas DataFrame.

df = pd.read_csv("retail_sales_dataset.csv")
```

```python
# 2.	Display the first 10 rows of the dataset.

df.head(10)
```

```python
# 3.	Check the total number of rows and columns.

df.shape
```

```python
# 4.	Display column names and their data types.

df.info()
```

```python
# 5.	Find the number of unique customers in the dataset.

df["Customer ID"].nunique()
```

```python
# 6.	Filter all transactions where Product Category is "Clothing".

clothing_Sales = df[df["Product Category"] == "Clothing"]

print(clothing_Sales)
```

```python
# 7.	Find all transactions where Total Amount is greater than 1000.

high_value_sales = df[df["Total Amount"] > 1000]
print(high_value_sales)
```

```python
# 8.	Calculate the total revenue generated from all sales.

df["Total Amount"].sum()

```

```python
# 9.	Calculate the average Price per Unit for each product category.

df.groupby("Product Category")["Price per Unit"].mean()
```

```python
# 10.	Find the total quantity sold for each product category.

df.groupby("Product Category")["Quantity"].sum()
```

```python
# 11.	Count the number of transactions made by each gender.

df["Gender"].value_counts()
```

```python
# 12.	Find the average Total Amount spent by male vs female customers.

df.groupby("Gender")["Total Amount"].mean()
```

```python
# 13.	Identify customers who have made more than one transaction.

customer_counts = df["Customer ID"].value_counts()
repeat_customers = customer_counts[customer_counts > 1]

print(repeat_customers)
```

```python
# 14.	Sort the dataset by Total Amount in descending order and show the top 5 rows.

df.sort_values("Total Amount", ascending=False).head(5)

```

```python
# 15.	Convert the Date column into datetime format.
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# 16.	Extract month and year from the Date column and add them as new columns.

df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

```python
# 17.	Calculate total monthly revenue.

monthly_revenue = df.groupby(["Year", "Month"])["Total Amount"].sum()
print(monthly_revenue)
```

```python
# 18.	Identify the month with the highest total sales.

monthly_revenue.idxmax(), monthly_revenue.max()
```

```python
# 19.	Find the age group (e.g., <25, 25–40, 40+) that contributes the most revenue.

bins = [0, 25, 40, 100]
labels = ["<25", "25-40", "40+"]

df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels)

df.groupby("Age Group")["Total Amount"].sum()
```

```python
# 20.	Verify whether Total Amount = Quantity × Price per Unit for all rows and identify mismatches, if any.

df["Calculated Amount"] = df["Quantity"] * df["Price per Unit"]

mismatch = df[df["Total Amount"] != df["Calculated Amount"]]
mismatch
```
