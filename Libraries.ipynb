```python
import math

print("\nðŸ“Œ Mathematical Constants")
print("pi =", math.pi)
print("e =", math.e)
print("tau =", math.tau)
print("inf =", math.inf)
print("nan =", math.nan)

print("\nðŸ“Œ Number Functions")
print("ceil(5) =", math.ceil(5))
print("floor(4.7) =", math.floor(4.7))
print("trunc(5) =", math.trunc(5))
print("fabs(-7.5) =", math.fabs(-7.5))
print("factorial(5) =", math.factorial(5))
print("gcd(12, 18) =", math.gcd(12, 18))
print("lcm(12, 18) =", math.lcm(12, 18))
print("isfinite(10) =", math.isfinite(10))
print("isinf(inf) =", math.isinf(math.inf))
print("isnan(nan) =", math.isnan(math.nan))

print("\nðŸ“Œ Power & Logarithmic Functions")
print("pow(2, 3) =", math.pow(2, 3))
print("sqrt(16) =", math.sqrt(16))
print("exp(2) =", math.exp(2))
print("expm1(2) =", math.expm1(2))
print("log(100, 10) =", math.log(100, 10))
print("log10(1000) =", math.log10(1000))
print("log2(8) =", math.log2(8))
print("log1p(9) =", math.log1p(9))

print("\nðŸ“Œ Hyperbolic Functions")
print("sinh(1) =", math.sinh(1))
print("cosh(1) =", math.cosh(1))
print("tanh(1) =", math.tanh(1))
print("asinh(1) =", math.asinh(1))
print("acosh(2) =", math.acosh(2))
print("atanh(0.5) =", math.atanh(0.5))

print("\nðŸ“Œ Special Functions")
print("comb(5, 2) =", math.comb(5, 2))
print("perm(5, 2) =", math.perm(5, 2))
print("dist([0,0], [3,4]) =", math.dist([0, 0], [3, 4]))
print("hypot(3, 4) =", math.hypot(3, 4))
print("prod([2,3,4]) =", math.prod([2, 3, 4]))
print("fsum([0.1]*10) =", math.fsum([0.1] * 10))
print("isclose(0.1+0.2, 0.3) =", math.isclose(0.1 + 0.2, 0.3))
```
