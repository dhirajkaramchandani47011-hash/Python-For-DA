```python
Purchase_Amount = int(input("Enter purchase amount: "))
Membership_Status = input("Enter membership status (None/Silver/Gold): ")

if Membership_Status == "None":
    print(f"Purchase amount is {Purchase_Amount}. No discount applied.")
elif Membership_Status == "Silver":
    Discount = Purchase_Amount * 0.05
    Final_Amount = Purchase_Amount - Discount
    print(f"Purchase amount is {Purchase_Amount}. Silver member discount applied (5%). Final amount is {Final_Amount}.")
elif Membership_Status == "Gold":
    Discount = Purchase_Amount * 0.10
    Final_Amount = Purchase_Amount - Discount - 100
    print(f"Purchase amount is {Purchase_Amount}. Gold member discount applied (10%) with extra 100 Rs Discount. Final amount is {Final_Amount}.")
```

```python
Side_A = int(input("Enter side A of the triangle: "))
Side_B = int(input("Enter side B of the triangle: "))
Side_C = int(input("Enter side C of the triangle: "))

if Side_A == Side_B and Side_B == Side_C:
    print("The triangle is Equilateral.")  
elif Side_A == Side_B or Side_B == Side_C or Side_A == Side_C:
    print("The triangle is Isosceles.")
elif Side_A != Side_B and Side_B != Side_C and Side_A != Side_C:
    print("The triangle is Scalene.")
else:
    print("Invalid input.")
    
```
