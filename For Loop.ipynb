```python
words = ["apple", "banana", "mango", "grapes"]

max_word = ""      
max_length = 0     

for word in words:
    if len(word) > max_length:
        max_length = len(word)
        max_word = word

print(f"The longest word is '{max_word}' with length {max_length}")
```

```python
s = input("Enter a string: ")
reversed_s = ""

for i in s:
    reversed_s = i + reversed_s   

print("Reversed string:", reversed_s)
```

```python
n = 10
total = 0

for i in range(1, n + 1):
    total += i ** 2   

print(f"Sum of squares from 1^2 to {n}^2 is {total}")
```

```python
nested_list = [[1, 2], [3, 4], [5]]
flat_list = []

for sublist in nested_list:
    for item in sublist:
        flat_list.append(item)

print("Flattened list:", flat_list)
```
